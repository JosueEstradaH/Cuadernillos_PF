# ----- Rehacer el arbol de decision -----
Como el arbol anterior cuenta con una precision de solamente 62.5%, agregaremos mas caracteristicas para obtener una mayor precision de prediccion. 

### 1. Importa la clase del modelo apropiado del Scikit Learn

```python
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
```

### 2. Escoge los hyperparámetros instanciando la clase con los valores deseados.

```python
R_decision_tree_metodos = DecisionTreeClassifier(random_state=0, criterion = "entropy")
```

### 3. Preparar los datos en matriz de características y vector de etiquetas.
En este nuevo arbol agregamos los datos de semestre, genero y carrera. Para poder utilizar las columnas de genero y carrera, tuvimos que modificar dichas columnas para que sus datos representen un valor numerico en vez de string. En el caso de genero, hombre=1 y mujer =0. Para las carreras, LICE=0, LISC=1, LGE=2, y LEI=3. 

```python
R_metodos_data_target_names=['video','pdf','audio','ejemplo']
R_metodos_data_feature_names=['Semestre','Genero','Carrera','Tiempo de elección','TiempoVideo',
                            'TiempoPDF','TiempoAudio','TiempoEjemplo'] ####

R_metodos_data = metodos_data 

R_genero= R_metodos_data.iloc[:, [3]] 
R_genero = R_genero.values

for x in range(96):
    if R_genero[x]=='Hombre':
        R_genero[x]=1
    else :
        R_genero[x]=0

R_carrera= R_metodos_data.iloc[:, [4]] 
R_carrera = R_carrera.values

for x in range(96):
    if R_carrera[x]=='LICE':
        R_carrera[x]=0
    if R_carrera[x]=='LISC':
        R_carrera[x]=1
    if R_carrera[x]=='LGE':
        R_carrera[x]=2
    if R_carrera[x]=='LEI':
        R_carrera[x]=3
```

```python

#columnas de las cuales queremos obtener datos
R_met_tiempo_data = R_metodos_data.iloc[:, [2,3,4,7, 8,9,10,11]] 

R_met_tiempo_data.iloc[:, [1]] = R_genero
R_met_tiempo_data.iloc[:, [2]] = R_carrera


# convertir esos datos a un array para trabajarlos con sklearn
R_met_tiempo_data = R_met_tiempo_data.values

#hacer lo mismo con el target 
R_met_target_data = R_metodos_data.iloc[:, [6]] 
R_met_target_data = R_met_target_data.values



print(type(metodos_data))
print(metodos_data.keys())

print(R_metodos_data_target_names)
print(R_metodos_data_feature_names)

print(R_met_tiempo_data)
print(R_met_target_data)
```

### 4. Ajustar el modelo a los datos llamando al método fit().

```python
R_decision_tree_metodos = R_decision_tree_metodos.fit(R_met_tiempo_data, R_met_target_data)
```

```python
R_decision_tree_metodos_text = export_text(R_decision_tree_metodos, feature_names=R_metodos_data_feature_names)
print(R_decision_tree_metodos_text)
```
